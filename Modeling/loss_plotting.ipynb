```python
import pandas as pd
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
with open('checkpoints/run_12GB_losses.txt') as f:
    data = [float(line.strip()) for line in f.readlines()]
len(data)
```

```python

data = pd.Series(data)
#reset index to reflect actual number of gradient steps
data.index = data.index * 350

#plot Loss over function of steps
fig = plt.figure(figsize=(10,6))
plot = data.plot()
ylabel=plt.ylabel('Loss', fontweight='bold', fontsize=14)
yticks = plt.yticks(fontweight='bold')
xlabel = plt.xlabel('Steps', fontweight='bold', fontsize=14)
title = plt.title('Model Loss as a function of Gradient Steps', fontweight='bold', fontsize=16)
```
