```python
import pandas as pd
import matplotlib.pyplot as plt
from math import floor
plt.style.use('ggplot')
```

```python
with open('checkpoints/run_4GB-wwm_losses.txt') as f:
    data = [float(line.strip()) for line in f.readlines()]
len(data)
```

```python

data = pd.Series(data)
#reset index to reflect actual number of gradient steps
data.index = data.index * 350 

#plot Loss over function of steps
fig = plt.figure(figsize=(10,6))
plot = data.plot()
ylabel=plt.ylabel('Loss', fontweight='bold', fontsize=14)
yticks = plt.yticks(fontweight='bold')
xlabel = plt.xlabel('Steps', fontweight='bold', fontsize=14)
title = plt.title('Model Loss as a function of Gradient Steps', fontweight='bold', fontsize=16)
```

```python
import numpy as np
```

```python
means = []

for x in range(floor(len(data)/10)):
    window = data.values[x * 10:(x+1)*10]
    mean = np.mean(window)
    means.append(mean)
[np.round(mean, 3) for mean in means]
```
