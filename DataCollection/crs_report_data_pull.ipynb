```python
page_count = 0
pdf_count = 0
errors = []

start = time.perf_counter()

for pdf in os.listdir():
    if os.path.isfile(pdf):
        if magic.from_file(pdf, mime=True) == 'application/pdf':
            pdf_count += 1
            if pdf_count % 1000 == 0:
                print(f'{pdf_count} docs counted, {len(os.listdir()) - pdf_count} to go.')
            try:
                #creates Document object from PDF
                doc = fitz.open(pdf)

                #process pdf page by page and join as one string
                pages = [page.get_text() for page in doc]
                page_count += len(pages)

            except Exception: 
                print(logging.exception(f'Found an error with this file {pdf}'))
                errors.append(pdf)
                continue
            
end = time.perf_counter() - start
print(f'Total time: {round(end,2)} seconds.')
print(f'Page count: {page_count}')
print(f'PDF count: {pdf_count}')
                        
```
